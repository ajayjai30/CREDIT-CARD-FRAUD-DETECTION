```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/drive/MyDrive/creditcard.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
g=StandardScaler()
```

```python
s=pd.DataFrame(df['Amount'])
```

```python
s=g.fit_transform(s)
```

```python
s=pd.DataFrame(data=s)
```

```python
df=df.drop("Amount",axis=1)
```

```python
df1=pd.concat([df,s],axis=1)
```

```python
df1.head()
```

```python
df1=df1.rename(columns={0:'NormalisedAmount'})
```

```python
df1
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
h=DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop("Class",axis=1)
y=df['Class']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
y_train=pd.DataFrame(y_train)
```

```python
h.fit(x_train,y_train)
```

```python
y_pred=h.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_pred,y_test)*100
```

```python
print(classification_report(y_pred,y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
e=RandomForestClassifier(n_estimators=500)
```

```python
e.fit(x_train,y_train)
```

```python
y_pred2=e.predict(x_test)
```

```python
accuracy_score(y_pred2,y_test)*100
```

```python
print(classification_report(y_pred2,y_test))
```

Out of both training models Randomforestclassifier performs better as it is an ensemble model
